#### The highest rated movies since 2000

```sql
SELECT Series_Title,
       Released_Year,
       IMDB_Rating
       FROM movies1
       WHERE Released_Year > 1999
       Order BY IMDB_Rating DESC
       LIMIT 10;
```

#### Best directors

###### Highest rated

```sql
Select Director,
       collect_list(Series_Title) AS Movies_list,
       array_size(collect_list(Series_Title)) AS Movies_number,
       ROUND(AVG(IMDB_Rating),1) AS Average_rating,
       ROUND(AVG(Gross),1) AS Average_gross /* a lot of lines missing in the original, replaced by average... not really telling */
       FROM movies1 
       GROUP BY Director
       HAVING Movies_number > 2
       ORDER BY average_rating DESC
       LIMIT 10; 
```

###### Hisgest grossing director (not 100% reliable due to 169 rows being null and replaced with 0)

```sql
Select Director,
       collect_list(Series_Title) AS Movies_list,
       array_size(collect_list(Series_Title)) AS Movies_number,
       ROUND(AVG(IMDB_Rating),1) AS Average_rating,
       ROUND(AVG(Gross),1) AS Average_gross
       FROM movies1 
       GROUP BY Director
       HAVING Movies_number > 2
       ORDER BY Average_gross DESC
       LIMIT 10; 
```

#### Actors with the highest movie ratings

```sql
SELECT Star,
       collect_list(Series_Title) AS Movies_list,
       ROUND(AVG(IMDB_Rating),1) AS Average_Rating,
       COUNT (*) AS Number_of_Movies
       FROM (
        SELECT Series_Title,
               explode(Stars) AS Star,
               IMDB_Rating
        FROM movies1
       )
       GROUP BY Star
       Having Number_of_Movies > 2
       ORDER BY Average_Rating DESC
       LIMIT 10
```

#### Most popular genres

```sql
SELECT Genres,
       COUNT(*) as Movies_count
      FROM movies1
      LATERAL VIEW explode(Genre) as Genres
      GROUP BY Genres
      ORDER BY Movies_count DESC
```
